```python
# load csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load data
data = pd.read_csv("datasets/salary.csv")
```

```python
# drop everything except the INDICATOR: Indicator, MUNICIPALITY: Municipality, TIME_PERIOD: Time period, and OBS_VALUE columns
data = data[["INDICATOR: Indicator", "MUNICIPALITY: Municipality", "TIME_PERIOD: Time period", "OBS_VALUE"]]
data
```

```python
# Get the rows with an indicator of "M010: Average"
averages = data[data["INDICATOR: Indicator"] == "M010: Average"]
# Get the 2 lowest rows with OBS_VALUE
lowest = averages.nsmallest(20, "OBS_VALUE")
lowest
```

```python
highest = averages.nlargest(10, "OBS_VALUE")
highest
```

```python
# Salary in Luxembourg
luxembourg = averages[averages["MUNICIPALITY: Municipality"] == "C580: Luxembourg"]
esch = averages[averages["MUNICIPALITY: Municipality"] == "C280: Esch-sur-Alzette"]
vianden = averages[averages["MUNICIPALITY: Municipality"] == "C990: Vianden"]
echternach = averages[averages["MUNICIPALITY: Municipality"] == "C240: Echternach"]

display(luxembourg)
display(esch)
display(vianden)
display(echternach)
```
